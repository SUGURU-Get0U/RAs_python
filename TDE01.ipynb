```python
# =======================================================================================
# Importacao de bibliotecas para o trablalho
import math
from IPython.display import display, Math # biblioteca para mostrar as equacoes!
import numpy as np
import matplotlib.pyplot as plt # Graficos
import sympy as sy




# EXERCICIO 2
'''
 2) Um terreno retangular tem área de 200 m² e deve ser cercado da seguinte maneira: dois lados opostos
devem receber uma cerca reforçada que custa R$ 300,00 o metro, e os dois lados restantes recebem uma
cerca padrão de R$ 200,00 o metro.
'''

# a) Modele uma função que expressa o custo total de cerca deste terreno?
display(Math(r'\text{a) Função que expressa o custo total (C) em função de x:}'))

# Declarando a variável simbólica x (comprimento de um dos lados)
x_sym = sy.symbols('x')

# Modelagem:
# Lados reforçados: 2 * 300 * x = 600x
# Lados padrão: 2 * 200 * (200/x) = 80000/x
custo_simbolico = 600 * x_sym + 80000 / x_sym

# Exibindo a fórmula formatada
display(Math(rf'C(x) = 600x + \frac{{80.000}}{{x}}'))
print(("==" * 30))

# b) Plote um gráfico que represente o custo da cerca em função do comprimento do terreno.
display(Math(r'\text{b) Gráfico do custo em função de x:}'))

# Criando um intervalo de valores para x (evitando o zero para não dar erro de divisão)
x_vals = np.linspace(1, 50, 400)
# Calculando o custo para cada valor de x
y_vals = 600 * x_vals + 80000 / x_vals

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, color="orange", linewidth=2.0)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('Comprimento do lado x (m)')
plt.ylabel('Custo Total (R$)')
plt.title("Custo da Cerca vs. Comprimento do Lado")
plt.show()
print(("==" * 30))

# c) Determine o limite do custo quando o x tende ao infinito.
display(Math(r'\text{c) Limite do custo quando } x \to \infty:'))
limite_inf = sy.limit(custo_simbolico, x_sym, sy.oo)

display(Math(rf'\lim_{{x \to \infty}} \left( 600x + \frac{{80.000}}{{x}} \right) = (600 * \infty + 0) = \infty'))
print("Interpretação: Se um lado for infinitamente comprido, o custo da cerca será infinito.")
print(("==" * 30))

# d) Determine o limite do custo quando o x tende a zero.
display(Math(r'\text{d) Limite do custo quando } x \to 0:'))
limite_zero = sy.limit(custo_simbolico, x_sym, 0)

display(Math(rf'\lim_{{x \to 0}} \left( 600x + \frac{{80.000}}{{x}} \right) = \infty'))
print("Interpretação: Se um lado for quase zero, o outro deve ser gigante para manter a área, tornando o custo infinito.")
print(("==" * 30))
```

```python
# =======================================================================================================
# Importacao de bibliotecas para o trablalho
import math
from IPython.display import display, Math # biblioteca para mostrar as equacoes!
import numpy as np
import matplotlib.pyplot as plt # Graficos
import sympy as sy

# =======================================================================================
# EXERCICIO 3
'''
3) A espessura da camada de creme formada sobre um café expresso na xícara, servido na cafeteria A,
no decorrer do tempo, é descrita pela função E(t) = a * 2^(bt), onde t >= 0 é o tempo (em segundos)
e a e b são números reais. Sabendo que inicialmente a espessura do creme é de 6 milímetros e que,
depois de 5 segundos, se reduziu em 50%.
'''

# a) Qual a espessura depois de 8 segundos?
display(Math(r'\text{a) Determinar a espessura } E(t) \text{ para } t = 8:'))

# --- PASSO 1: Encontrar 'a' (Espessura inicial t=0) ---
display(Math(r'E(0) = a \cdot 2^{b \cdot 0} = a \cdot 1 = 6'))
a = 6

# --- PASSO 2: Encontrar 'b' (Usando t=5 e redução de 50%) ---
# Se reduziu 50%, a nova espessura é 3mm --> 6 / 2.
# 3 = 6 * 2^(b * 5)
display(Math(r'0.5 = 2^{5 \cdot b} = 2^{-1} = 2^{5 \cdot b} = -1 = 5b -> b = -1/5'))
b = -1/5

# --- PASSO 3: Calcular para t = 8 ---
t_alvo = 8
espessura_8s = a * (2**(b * t_alvo))

# Exibição Matemática
display(Math(r'E(t) = 6 \cdot 2^{-1/5 * 8}'))
display(Math(rf'E(8) = 6 \cdot 2^{{-1/5 * 8}} \approx {espessura_8s:.2f}mm'))
print(("==" * 30))

# b) Plote o gráfico da espessura em função do tempo.
# Definindo a função baseada nos seus cálculos
def E(t):
    return 6 * (2**(-1/5 * t))

# Criando os dados para a curva (de 0 a 15 segundos)
t_espaco = np.linspace(0, 15, 100)
e_valores = E(t_espaco)

# Criando a figura
plt.figure(figsize=(10, 6))

# Plotando a linha do decaimento
plt.plot(t_espaco, e_valores, color='#6F4E37', linewidth=3, label=r'$E(t) = 6 \cdot 2^{-1/5 \cdot t}$')

# Destacando os pontos específicos que você calculou
pontos_t = [0, 5, 8]
pontos_e = [E(0), E(5), E(8)]
labels = ['Início (6mm)', '5s (3mm)', f'8s ({E(8):.2f}mm)']

for i in range(len(pontos_t)):
    # Desenha o ponto vermelho
    plt.scatter(pontos_t[i], pontos_e[i], color='red', zorder=5)
    # Adiciona o rótulo ao lado do ponto
    plt.annotate(labels[i], (pontos_t[i], pontos_e[i]),
                 textcoords="offset points", xytext=(10,5), ha='left', fontsize=10)

# Linhas
plt.title('Decaimento da Espessura do Creme no Café Expresso', fontsize=14)
plt.xlabel('Tempo (segundos)', fontsize=12)
plt.ylabel('Espessura (mm)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()

# Mostrando o gráfico
plt.show()

# c) Analise o comportamento da função quando o tempo for muito grande, tender ao infinito. Interprete o
# resultado no contexto físico do problema.
```

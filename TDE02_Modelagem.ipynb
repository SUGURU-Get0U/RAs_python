```python
# ==================================================================
# -- EXERCICIO 01 --
# ==================================================================

# Importação de bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
from IPython.display import display, Math

# Variáveis simbólicas globais
x_sym, t_sym, n_sym, a_sym = sy.symbols('x t n a')

print("="*80)
print("                                 RESOLUÇÃO TDE 02")
print("="*80)

# =======================================================================================
# EXERCICIO 1
# =======================================================================================
display(Math(r'\text{1) A curva } y = \frac{1}{1+x^2} \text{ (Bruxa de Maria Agnesi)}'))

# a) Derivada
display(Math(r'\text{a) Calcule } \frac{dy}{dx}:'))
y_expr = 1 / (1 + x_sym**2)
dy_dx = sy.diff(y_expr, x_sym)
display(Math(rf'\frac{{dy}}{{dx}} = {sy.latex(dy_dx)}'))

# b) Reta Tangente
display(Math(r'\text{b) Equação da reta tangente no ponto } P(-1, 1/2):'))
m = dy_dx.subs(x_sym, -1)
# y - 1/2 = m(x + 1)
reta_expr = m * (x_sym + 1) + 0.5
display(Math(rf'm = f\'(-1) = {m}'))
display(Math(rf'y - 0.5 = {m}(x + 1) \implies y = {sy.latex(sy.simplify(reta_expr))}'))

# c) Gráficos
x_vals = np.linspace(-5, 5, 400)
f_y = lambda x: 1 / (1 + x**2)
f_t = lambda x: 0.5 * x + 1.0

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(x_vals, f_y(x_vals), label='Curva')
ax1.plot(x_vals, f_t(x_vals), '--', label='Tangente')
ax1.scatter([-1], [0.5], color='red')
ax1.set_title("Geral")
ax1.grid(True)

ax2.plot(x_vals, f_y(x_vals))
ax2.plot(x_vals, f_t(x_vals), '--')
ax2.set_xlim(-1.1, -0.9)
ax2.set_ylim(0.45, 0.55)
ax2.set_title("Zoom no Ponto P")
ax2.grid(True)
plt.show()
print("==" * 40)
```

```python
# =======================================================================================
# EXERCICIO 2
# =======================================================================================

import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
from IPython.display import display, Math

display(Math(r'\text{2) Encontre } f\'(x):'))

itens = [
    (3*x_sym**2 - 12/x_sym + sy.sqrt(x_sym), "a"),
    ((x_sym**2 + 3*x_sym)/sy.pi, "b"),
    ((2*x_sym - 1)*(x_sym**3 + 4*x_sym), "c"),
    (x_sym**(sy.Rational(1,3)) / (x_sym - 3), "d"),
    ((sy.Symbol('a')*x_sym + sy.Symbol('b'))/(sy.Symbol('c')*x_sym + sy.Symbol('d')), "e")
]

for expr, letra in itens:
    deriv = sy.diff(expr, x_sym)
    display(Math(rf'\text{{{letra}) }} f(x) = {sy.latex(expr)} \implies f\'(x) = {sy.latex(sy.simplify(deriv))}'))

print("==" * 40)
```

```python
# =======================================================================================
# EXERCICIO 3
# =======================================================================================

import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
from IPython.display import display, Math

display(Math(r'\text{3) Análise de Receita } R(p) = p \cdot f(p)'))

display(Math(r'\text{a) Interpretação de } f(20)=10000 \text{ e } f\'(20)=-350:'))
print("Significa que ao preço de R$ 20,00, a demanda é de 10.000 unidades.")
print("A taxa de variação indica que, se o preço aumentar R$ 1,00, a demanda cai em 350 unidades.")

display(Math(r'\text{b) Cálculo de } R\'(20):'))
# R'(p) = f(p) + p*f'(p)
r_p = 10000 + 20 * (-350)
display(Math(rf'R\'(20) = f(20) + 20 \cdot f\'(20) = 10000 - 7000 = {r_p}'))
print(f"Interpretação: A receita está aumentando à taxa de R$ {r_p} por real de aumento no preço.")
print("==" * 40)
```

```python
# =======================================================================================
# EXERCICIO 4
# =======================================================================================
display(Math(r'\text{4) Projétil: } h(t) = 2 + 24.5t - 4.9t^2'))

h_t = 2 + 24.5*t_sym - 4.9*t_sym**2
v_t = sy.diff(h_t, t_sym)

display(Math(rf'\text{{a) Velocidade: }} v(t) = {sy.latex(v_t)}'))
display(Math(rf'\text{{b) }} v(2) = {v_t.subs(t_sym, 2)} \text{{ e }} v(4) = {v_t.subs(t_sym, 4)}'))

t_subida = sy.solve(v_t, t_sym)[0]
h_max = h_t.subs(t_sym, t_subida)
display(Math(rf'\text{{c) Altura Máxima em }} t = {t_subida}s \text{{ é }} h = {h_max}m'))

# Solo
t_solo = sy.solve(h_t, t_sym)[1] # raiz positiva
display(Math(rf'\text{{e) Atinge o solo em }} t \approx {t_solo:.2f}s'))
display(Math(rf'\text{{f) Velocidade de impacto: }} v({t_solo:.2f}) \approx {v_t.subs(t_sym, t_solo):.2f} m/s'))

# Gráfico Trajetória
t_p = np.linspace(0, float(t_solo), 100)
h_p = [float(h_t.subs(t_sym, val)) for val in t_p]
plt.figure(figsize=(6,3))
plt.plot(t_p, h_p, color='purple')
plt.scatter([float(t_subida)], [float(h_max)], color='red')
plt.title("Trajetória do Projétil")
plt.xlabel("Tempo (s)"); plt.ylabel("Altura (m)")
plt.grid(True)
plt.show()
print("==" * 40)
```

```python
# =======================================================================================
# EXERCICIO 5
# =======================================================================================
display(Math(r'\text{5) Otimização de Volume (Caixa sem tampa)}'))

# Area = a^2 + 4ah = 1200 -> h = (1200 - a^2) / 4a
# V = a^2 * h = a(1200 - a^2)/4
v_caixa = (300 * a_sym) - (a_sym**3 / 4)
display(Math(rf'\text{{a) Função Volume: }} V(a) = {sy.latex(v_caixa)}'))

dv_da = sy.diff(v_caixa, a_sym)
a_critico = sy.solve(dv_da, a_sym)[1] # raiz positiva
v_final = v_caixa.subs(a_sym, a_critico)
display(Math(rf'\text{{b) Volume máximo: }} a = {a_critico} \implies V = {v_final} cm^3'))

print("==" * 40)
```

```python
# =======================================================================================
# EXERCICIO 6
# =======================================================================================
display(Math(r'\text{6) Minimização de Tempo (Cluster)}'))

tempo_func = 400/n_sym + 0.25*n_sym
dt_dn = sy.diff(tempo_func, n_sym)
n_otimo = sy.solve(dt_dn, n_sym)[1]
display(Math(rf'T\'(n) = {sy.latex(dt_dn)} = 0 \implies n = {n_otimo}'))
print(f"O número ideal é de {n_otimo} processadores.")
print("==" * 40)
```

```python
# =======================================================================================
# EXERCICIO 7
# =======================================================================================
display(Math(r'\text{7) Cálculo de Integrais}'))

ints = [
    (3*x_sym**2 + 12*x_sym - 5, "a", None),
    ((2 + sy.sqrt(x_sym))**2, "b", None),
    (x_sym**(sy.Rational(1,3)), "c", (1, 8)),
    ((x_sym**4 + 3*x_sym**6)/x_sym**4, "d", (1, 2))
]

for expr, letra, limites in ints:
    if limites:
        res = sy.integrate(expr, (x_sym, limites[0], limites[1]))
        display(Math(rf'\text{{{letra}) }} \int_{{{limites[0]}}}^{{{limites[1]}}} {sy.latex(expr)} \, dx = {sy.latex(res)}'))
    else:
        res = sy.integrate(expr, x_sym)
        display(Math(rf'\text{{{letra}) }} \int {sy.latex(expr)} \, dx = {sy.latex(res)} + C'))

print("==" * 40)
```

```python
# =======================================================================================
# EXERCICIO 8
# =======================================================================================
display(Math(r'\text{8) Partícula: } a(t) = t + 4, v(0) = 5'))

v_part = sy.integrate(t_sym + 4, t_sym) + 5
display(Math(rf'\text{{a) Velocidade: }} v(t) = {sy.latex(v_part)}'))

distancia = sy.integrate(v_part, (t_sym, 0, 10))
display(Math(rf'\text{{b) Distância (0 a 10s): }} \int_{{0}}^{{{10}}} v(t) \, dt = {distancia.evalf():.2f} m'))

print("="*80)
print("                               FIM DO TRABALHO")
print("="*80)
```
